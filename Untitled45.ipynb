```python
X = df.loc[:,['income','spending_score']].values
```

```python
print("\n Calculating WCSS fot the Elbow Method....")
wcss=[]
for i in range(1,11):
  Kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42, n_init = 10)
  Kmeans.fit(X)
  wcss.append(Kmeans.inertia_)
```

printing elbow Method graph

```python
plt.figure(figsize = (12,6))
plt.plot(range( 1,11),wcss,marker ='o', linestyle='--')
plt.title('The Elbow method',fontsize = 19)
plt.xlabel('Number of Clusters ( k)', fontsize = 12)
plt.ylabel('WCSS(Within-Cluster Sum of Squares)',fontsize = 12)
plt.xticks(range(1,11))
plt.grid(True)
plt.show()
```

```python
print ("the optimal number of clusters in 3")
```

Apply K-Means with the Optimal Number of Clusters (K=5)

```python
print("\n Training the K-Means Model with K = 3....")
kmeans = KMeans(n_clusters= 3, init= 'k-means++', random_state =42, n_init= 10)
```

```python
y_kmeans = kmeans.fit_predict(X)
```

Silhouette score

```python
silhouette = silhouette_score(X, y_kmeans)
print(f"Silhouette Score: {silhouette:.2f}")
print("score is between -1 and 1. Higher is better. It measures how simmilar a data point is to its own cluster compared to other cluster.")
```
